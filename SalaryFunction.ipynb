```python
import pandas as pd
import zipfile
import os
```

```python
# Import salary data
data_path = r"C:\Users\OlannaOgbenna\Documents\NXU"
try:
    df = pd.read_csv(data_path)
    print("Total.csv loaded successfully.")
except Exception as e:
    print(f"Error loading data: {e}")
    df = pd.DataFrame()  # empty DataFrame if load fails
```

```python
# Convert DataFrame to a dictionary keyed by employee name
employees = {}
if not df.empty:
    employees = df.set_index('EmployeeName').T.to_dict()
    print("Employee dictionary created.")
else:
    print("DataFrame is empty; cannot create employee dictionary.")
```

```python
# Define function to get employee details
def get_employee_details(name):
    """Return details for an employee by name, or None if not found."""
    try:
        if name in employees:
            return employees[name]
        else:
            raise KeyError(f"Employee '{name}' not found.")
    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
# Example usage: get details for a specific employee
employee_name = "NATHANIEL FORD"
result = get_employee_details(employee_name)
if result:
    print(f"Details for {employee_name}: {result}")
else:
    print(f"No details found for '{employee_name}'.")
```

```python
# Export employee details to CSV and zip the CSV
if result:
    details_df = pd.DataFrame([{**{'Name': employee_name}, **result}])
    csv_filename = f"{employee_name}_details.csv"
    output_dir = r"C:/Users/OlannaOgbenna/WorkerPayslip"
    output_csv_path = os.path.join(output_dir, csv_filename)

     # Write to CSV
    try:
        details_df.to_csv(output_csv_path, index=False)
        print(f"Wrote employee details to {output_csv_path}.")
    except Exception as e:
        print(f"Error writing CSV: {e}")
        
         # Create a zip file and write the CSV into it
    zip_path = os.path.join(output_dir, "Employee Profile.zip")
    try:
        with zipfile.ZipFile(zip_path, mode="w") as archive:
            archive.write(output_csv_path, arcname=csv_filename)
        print(f"Created ZIP archive at {zip_path}.")
    except Exception as e:
        print(f"Error creating zip archive: {e}")
```
